```julia
#we work with A_new_b and d_new_b and uper_new_b that have smaller number of rows compare to part a
v = sparse(ones(8763))
measures=A*v
sum(measures)/362462

q=0
for i in 1:362462
    if measures[i]>=20 && d[i]==2 #harder than a:
        q += 1
    end
end
q
```

```julia
using Convex, SCS
x = Variable(8763)
problem_b = minimize(norm(A_new_b*x-d_new_b, 2))
problem_b.constraints += [x >= 0; A_new_b*x <= uper_new_b]
```
